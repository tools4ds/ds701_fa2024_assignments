# Introduction to Networks

## Graphs allow us to represent and analyze complex relationships and structures in data

### - A graph G = (V, E) is a pair, where V is a set of vertices, and E is a set of unordered vertex pairs called edges.

### - The term nodes is also used for vertices. The term links or connections is also used for edges.



## Undirected Graphs

### In an undirected graph, an edge (u, v) is an unordered pair. The edge (v, u) is the same thing.



## Directed Graphs

### - In an directed graph, an edge (u, v) is an ordered pair and is different from the edge (v, u). 
### - The edges have an orientation.



## Path

### A path in a graph from node u to node v is a sequence of edges that starts at u and ends at v. Paths exist in both undirected and directed graphs.



## Degree 

### The degree of a node is the number of edges that connect to it.



## Let's implement some of these concepts on Python and visualize them.

### 1. Random Graphs Model: G(n, p) 

- The G (n, p) random graph model generates a graph with n nodes where each edge is included with probability p. 
- This is useful for simulating random networks.

```python
import networkx as nx
import matplotlib.pyplot as plt

n = 10  # Number of nodes
p = 0.3  # Probability of edge creation

#we can generate a random graph like this
G = nx.gnp_random_graph(n, p)

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True)
plt.show()
```

## 2. Degree Distributions

- The degree distribution shows `how many nodes` have each `possible degree` (number of edges). 
- `Power-law distributions` are common in real-world networks, such as `social networks`.

```python
import numpy as np

#let's generate a random graph (or use any other graph)
G = nx.gnp_random_graph(20, 0.8)

plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True)
plt.show()
```

```python
#get degrees of all nodes
degree_sequence = [d for n, d in G.degree()]

plt.hist(degree_sequence, bins=np.arange(min(degree_sequence), max(degree_sequence) + 1), alpha=1)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

## 3. Clustering Coefficients

- The clustering coefficients helps us measure `how well connected` a `node's neighbors` are to each other.
- Useful for understanding the `local structure` of networks.


```python
#calculate clustering coefficients using nx.clustering
clustering_coefficients = nx.clustering(G)

#average clustering coefficient
avg_clustering = sum(clustering_coefficients.values()) / len(clustering_coefficients)
print(f"Clustering Coefficients: {clustering_coefficients}")
print(f"Average Clustering Coefficient: {avg_clustering}")
```

## 4. Small World Networks

- `Small-world networks` have `short path lengths` and `high clustering coefficients`. 
- The `Watts-Strogatz model` can generate small-world graphs.

```python
n = 30  #number of nodes
k = 4  #each node is connected to k nearest neighbors in ring topology
p = 0.3 #probability of rewiring each edge

#generate small-world network using watts_strogatz
G_sw = nx.watts_strogatz_graph(n, k, p)

#draw the small-world network
plt.figure(figsize=(8, 6))
nx.draw(G_sw, with_labels=True)
plt.show()
```

## 5. Visualizing layouts of different structures

- Visualizing graphs is `crucial for understanding` their structure. 
- `Different layouts` like `spring layout` and `spectral layout` can be used to `visualize graphs` in Python.

```python
G = nx.gnp_random_graph(20, 0.8)
```

### The spring layout positions nodes using a `force-directed algorithm`

```python
pos = nx.spring_layout(G)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True)
plt.title("Spring Layout")
plt.show()
```

### The spectral layout uses `eigenvectors of the Laplacian matrix` to position nodes.

```python
pos_spectral = nx.spectral_layout(G)

plt.figure(figsize=(8, 6))
nx.draw(G, pos_spectral, with_labels=True)
plt.title("Spectral Layout")
plt.show()

```
